# 디지털정보지수 (5,6,70대)

* 추출방법 : 디지털 정보지수는 **[2021 연령대별 디지털 역량 지수] * [행정동별 연령대별 인구 수]** 로 계산해 생성
* 사용이유 : 행정동별 시니어분들의 디지털 이용 수준을 알고자 이용

```python
media = pd.read_excel('/content/drive/MyDrive/공유폴더/완성데이터/원천 데이터(데이터 가공 전)/동별_연령대별_인구_from 서울시열린데이터.xlsx')
media[:4]
```

```python
# 불필요한 컬럼 삭제 
media.drop(columns = ['0~4세', '5~9세', '10~14세', '15~19세',
                     '20~24세', '25~29세','30~34세', '35~39세','40~44세', '45~49세',
                      '50~54세', '55~59세',  '60~64세', '65~69세','70~74세', '75~79세', '80~84세', '85~89세'], inplace = True)
```

```python
# 정리한 파일을 1차 저장
media.to_csv('동별_연령별_인원수for미디어정보지수.csv')
```

```python
#50대, 50대, 70대 이상의 역량수준
dgt_level = [0.92,0.593,0.224]


#역량수준 * 해당 연령대 인구 수
result = round(media.iloc[:,5:] * dgt_level,2)
result
```

```python
#media의 행정동 column과 결과값을 합쳐 하나의 DF로 만들기 
media_df = pd.concat([media[['동']],result], axis = 1)

#각 연령대별 디지털정보화 총합 구하기 
media_df['디지털정보지수'] = media_df.sum(axis =1)
media_df
```

```python
#행정동과 디지털정보지수만 저장 
final_df = media_df[['동','디지털정보지수']]
final_df
```

```python
final_df.to_csv('동별 디지털정보지수(50~).csv')
final_df.to_excel('동별 디지털정보지수(50~).xlsx')
```
