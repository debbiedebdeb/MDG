```python
from geopy.geocoders import Nominatim 
```

```python
df = pd.read_csv('/content/drive/MyDrive/공유폴더/완성데이터/NEW(공모전용)/연면적.csv')
df.head(1)
```

```python
geolocoder = Nominatim(user_agent = 'South Korea', timeout = None)

def geocoding(address) : 
    try : 
        geo= geolocoder.geocode(address)
        crd = (str(geo.latitude), str(geo.longitude))
        return crd
    
    except : 
        return (0,0)
```

```python
lats = []
logs = []

for address in df['주소'] : 
    crd = geocoding(address)
    lats.append(crd[0])
    logs.append(crd[1])

lats = pd.Series(lats)
logs = pd.Series(logs)
```

```python
df = pd.concat([df, lats], axis = 1)
df = pd.concat([df, logs], axis = 1)
df.head(5)
```

```python
df.columns = ['자치구', '동', '시설명', '주소', '연면적', '삭제1','삭제2', '위도', '경도', 
              '삭제3', '삭제4']
df.head(1)
```

```python
!pip install haversine

from haversine import haversine
```

```python
df = pd.read_excel('/content/drive/MyDrive/공유폴더/NEW(공모전용)/데이터위도경도포함.xlsx')
df.head(1)
```

```python
df_bus = pd.read_csv('/content/drive/MyDrive/공유폴더/EDA 및 탐색/Data 수집/주변특성/서울시버스정류소좌표데이터(2022.04.15).csv', 
                     encoding ='cp949')
df_bus
```

```python
#df_bus (경도 위도)를 위도 경도로  바꿔준다 
df_bus[['ARS-ID', '정류소명', 'Y좌표', 'X좌표']]
```

```python
from tqdm import tqdm

# 함수만들기
def find_bus_cnt(center, store, column) :
    for i in tqdm(range(len(center))) :
        distance = []
        for j in range(len(store)) : 
            centerloc = tuple(df[['위도', '경도']].iloc[i])
            storeloc = tuple(df_bus[['Y좌표', 'X좌표']].iloc[j])
            distance.append(haversine(centerloc, storeloc, unit = 'm'))
            
        distance = [d for d in distance if d <= 200.0]
        center[column][i] = len(distance)
        
    return center
```

```python
df['200m내버스정류장개수'] = 0 
df = find_bus_cnt(df, df_bus,'200m내버스정류장개수')
df
```
